```python
def chapter1():
    aakkoset = {"A": ("auto", "car"), "B": ("banaani", "banana"), "C": ("cheddar", "cheddar cheese"),
                "D": ("dekkari", "detective story"), "E": ("ei", "no"), "F": ("faarao", "pharaoh"),
                "G": ("galaksi", "galaxy"), "H": ("hieno", "fine"), "I": ("ilo", "pleasure/joy"),
                "J": ("jalka", "foot"), "K": ("kauppa", "shop"), "L": ("lasi", "glass"), "M": ("mummo", "grandma"),
                "N": ("nainen", "woman"), "O": ("omena", "apple"), "P": ("poika", "boy"),
                "Q": ("quo vadis", "where are you going?"), "R": ("raha", "money"), "S": ("silmä", "eye"),
                "T": ("täti", "aunt"), "U": ("uusi", "new"), "V": ("vuori", "mountain"), "W": ("WC", "WC"),
                "X": ("xylitoli", "xylitol"), "Y": ("yksin", "alone"), "Z": ("zoomata", "zoom"),
                "Å": ("Åland", "Åland"), "Ä": ("äiti", "mother"), "Ö": ("ötökkä", "bug/noise")}
    
    for key, value in aakkoset.items():
    
        if key == "Q" or key == "X":
            print(f"{key}\t{value[0]}\t{value[1]}")
        else:
            print(f"{key}\t{value[0]}\t\t{value[1]}")
    
    print("\nSpecial Characters - Å, Ä, Ö\n")
    print("Å is only found in a few words coming directly from swedish.\n\nÅland (Ahvenanmaa)\nÅke (a man's name)\n")
    print("Å is pronounced as a regular O.\n")
    
    print("Ä is the same sound as in the english word cat or hand.\n\näla (don't)\näkkiä (suddenly)\n")

    print("Ö is the same sound as in the English word learn or search.\n\nöljy (oil)\nyö (night)\n")

    return
```

```python
def chapter6():
    family_members = {"mother": ("äiti", "äippä", "äiskä"), "father": ("isä", "isi", "iskä"),
                      "sibling": ("sisarus",), "sister": ("sisko", "sisar"), "brother": ("veli", "velipoika"),
                      "grandmother": ("isoäiti", "mummo", "mummi", "mummu"),
                      "grandfather": ("isoisä", "vaari", "pappa", "ukki"), "aunt": ("täti",),
                      "uncle": ("setä (father's brother)", "eno (mother's brother)"), "cousin": ("serkku",),
                      "stepmother": ("äitipuoli",), "stepfather": ("isäpuoli",),
                      "half sister / stepsister": ("siskopuoli",), "half brother / stepbrother": ("velipuoli",)}
    
    note = u"\n\u27F6 words for Mom and Dad aren't capitalised in Finnish (except when they begin a sentence)\n"
    
    for member in family_members:
        values = " / ".join(family_members[member])
        print(f"{member:<40}{values}")
        if member == "brother":
            print(note)

    print(f"\n{'I have...':<40}Minulla on...\n\nFor example:\n{'I have a father and a sister.':<40}Minulla on isä ja sisko.\n")
    print("Plural is formed with the suffixes -a or -ä:")
    
    plural_forms = {"two sisters": ("kaksi siskoa",), "two grandmas": ("kaksi mummoa", "mummia", "mummua", "isoäitiä"),
                    "two granddads": ("kaksi vaaria", "ukkia", "pappaa", "isoisää"), "three brothers": ("kolme veljeä",),
                    "two dads": ("kaksi isää",), "two moms": ("kaksi äitiä",)}
    
    note2 = u"\n\u27F6 Don't stress about getting the suffix right, Finns will understand you even if you get it wrong!"
    
    for form in plural_forms:
        values = " / ".join(plural_forms[form])
        print(f"{form:<40}{values}")
    
    print(note2)
    
    return
```

```python
def chapter7():
    värit = {"blue":"sininen", "red":"punainen", "green":"vihreä", "yellow": "keltainen", "orange": "oranssi",
            "grey / gray": "harmaa", "brown": "ruskea", "purple": ("violetti", "liila"), "black": "musta",
            "white": "valkoinen"}

    for color in värit:
        if type(värit[color]) is tuple:
            values = " / ".join(värit[color])
            print(f"{color:<40}{values}")
        else:
            print(f"{color:<40}{värit[color]}")

    print("Some examples of use:\n")

    examples = {"a black cat": "musta kissa", "a red car": "punainen auto", "a green bag": "vihreä laukku",
               "a white dog": "valkoinen koira", "a purple toy": ("violetti", "liila lelu")}

    for ex in examples:
        if type(examples[ex]) is tuple:
            values = " / ".join(examples[ex])
            print(f"{ex:<40}{values}")
        else:
            print(f"{ex:<40}{examples[ex]}")

    print(u"\n\u27F6 Finnish doesn't have articles (a/an/the)!")

    print("\nMore colors - Lisää värejä\n")

    more_colors = {"light blue": "vaaleansininen", "light green": "vaaleanvihreä", "light": "vaalean",
                  "pink": ("vaaleanpunainen", "pinkki"), "dark red": "tummanpunainen",
                   "dark green": "tummanvihreä", "dark": "tumman"}

    for more in more_colors:
        if more == "light" or more == "dark":
             print(f"\n\u27F6 {more:<40}{more_colors[more]}\n")
        elif more == "pink":
            values = " / ".join(more_colors[more])
            print(f"Note:\n{more:<40}{values}\n")
        else:
            print(f"{more:<40}{more_colors[more]}")
    
    return
```

```python
def quiz_scores(total, rounds=0):
    if rounds == 0:
        if total <= 2:
            return "\nYou can do better than this :|"
        elif total <= 4:
            return "\nBetter luck next time!"
        elif total == 5:
            return "\nYou got half of the questions correct!"
        elif total <= 7:
            return "\nOk, you're getting the hang of this!"
        elif total <= 9:
            return "\nWow, impressive score!"
        elif total == 10:
            return "\nYou dropped your crown your highness :O Perfect score!!!"
    else:
        number_questions = rounds * 10
        if total <= number_questions/10:
            return "\nWell, this quiz didn't go very well :/ There's always a next time!"
        elif total <= number_questions/5:
            return "\nIt's only up from here! Better hit those books ;)"
        elif total <= number_questions/2.5:
            return "\nYou almost got half of the questions correct! Keep going!"
        elif total <= number_questions/1.6666666:
            return "\nYou're warming up! A little more practice and you'll burn the house down ;)"
        elif total <= number_questions/1.25:
            return "\nGreat score! I can see you've been practicing :)"
        elif total < number_questions:
            return "\nLet's go! Your score is almost as big as Trump's tariffs ;) \n"
        else:
            return f"\nWow, astonishing result :D {number_questions}/{number_questions}!!! Can I be like you when I grow up?"
```

```python
def get_three_letters(str):
    lst_words = []
    comparison = min(3, len(str))
    while len(lst_words) < comparison:
        rand_ind = random.randint(0, len(str) - 1)
        letter = str[rand_ind]
        if letter.isalpha() == True and letter not in lst_words:
            lst_words.append(letter.lower())
    return lst_words
```

```python
def ask_question(question_type, key, vocab, list_keys, proverbs=False):
    correct = 0
    count = 0
    poss_words = random.sample(list_keys, 4)
    if key not in poss_words:
        poss_words[3] = key
    random.shuffle(poss_words)
    
    if proverbs == False:
        word = 'word'
        value = vocab[key][0]
        list_values = vocab[key]
        values = [vocab[w][0] for w in poss_words]
    else:
        word = 'proverb'
        value = vocab[key]
        list_values = (vocab[key].lower(),)
        values = [vocab[w] for w in poss_words]
        
    if question_type == 0:
        print(f"What does the {word} {value} mean in english?")
        answer = input("")
        while answer.lower() != key.lower():
            if count == 0:
                letters = ", ".join(get_three_letters(key))
                print(f"The {word} has the letters {letters}")
            elif count == 3:
                print("Let me help you out. Choose one of the following options:")
                for w in poss_words:
                    print(f"- {w}")
            elif count == 6:
                print(f"The right answer is {key}. You'll get it next time!")
                break
            answer = input("Try to guess again, please:\n")
            count += 1
        if answer.lower() == key.lower():
            correct = 1
            print("Great! Your answer is correct :D")
    else:
        print(f"What does the {word} {key} mean in finnish?")
        answer = input("")
        while answer.lower() not in list_values:
            if count == 0:
                letters = ", ".join(get_three_letters(value))
                print(f"The {word} has the letters {letters}")
            elif count == 3:
                print("Let me help you out. Choose one of the following options:")
                for w in values:
                    print(f"- {w}")
            elif count == 6:
                print(f"The right answer is {value}. You'll get it next time!")
                break
            answer = input("Try to guess again, please:\n")
            count += 1
        if answer.lower() in list_values:
            correct = 1
            print("Great! Your answer is correct :D")
            
    return correct
```

```python
import random
import numpy as np

def ask_random(vocab, proverb=False):
    word = 'vocabulary'
    max_rounds = int(len(vocab)/10)
    choice1 = input(f"This quiz asks 10 questions in each round. How many rounds do you want to do? (Choose a number between 1 and {max_rounds}): ")
    count = 0

    while True:
        if choice1.isdigit() == True:
            if int(choice1) not in list(range(1, max_rounds+1)):
                if count == 4:
                    print("Since your response does not follow the criteria, the quiz will end.\nGoodbye! Until next time :)")
                    return
                choice1 = input(f"Please choose a number between 1 and {max_rounds}: ")
                count += 1
            else:
                break
        else:
            if count == 4:
                print("Since your response does not follow the criteria, the quiz will end.\nGoodbye! Until next time :)")
                return
            choice1 = input(f"You didn't input a number :/ Please choose a number between 1 and {max_rounds}: ")
            count += 1
        
    if proverb == True:
        word = 'proverbs'

    count = 0
    choice3 = input(f"Would you like the {word} to be displayed before the quiz? Please answer yes or no: ")
    while (choice3.lower() != "no" and choice3.lower() != "n" and choice3.lower() != "yes" and choice3.lower() != "y"):
        if count == 3:
            print("Since your response does not follow the criteria, the quiz will end.\nGoodbye! Until next time :)")
            return
        choice3 = input("Please respond with yes or no: ")
        count += 1
    list_keys = list(vocab.keys())
    random.shuffle(list_keys)
    total_score = 0
    for n in range(int(choice1)):
        sublist_keys = list_keys[n*10:(n+1)*10]
        print(f"\nRound {n+1}\n")
        if choice3.lower() == "yes" or choice3.lower() == "y":
            for key in sublist_keys:
                if proverb == False:
                    if len(vocab[key]) > 1:
                        values = ", ".join(vocab[key])
                    else:
                        values = vocab[key][0]
                    print(f"{key:<25}{values}")
                else:
                    values = vocab[key]
                    print(f"{key:<65}{values}")
            choice4 = input("\nWhen you're ready to move on press enter: ")
        # question_type contains the type of question
        # if it's 0, it asks for the english translation of a finnish word/proverb
        # if it's 1, it asks for the finnish translation of an english word/proverb
        question_type = np.random.randint(2, size=10)
        quiz_score = 0
        for index, key in enumerate(sublist_keys):
            question_score = ask_question(question_type[index], key, vocab, list_keys, proverbs=proverb)
            quiz_score += question_score
            total_score += question_score
        print(f"\nYou scored {quiz_score}/10")
        print(quiz_scores(quiz_score) + "\n")
    print(f"\nCongratulations, you finished the quiz! Your score is {total_score}/{total_score}")
    print(quiz_scores(total_score, rounds=int(choice1)))
```
